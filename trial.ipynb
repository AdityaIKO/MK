```python
import pandas as pd
```

```python
# Replace 'file_path.xlsx' with the path to your Excel file
# Load all sheets into a dictionary of DataFrames
df_dict = pd.read_excel('Penjualan MK 2.xlsx', sheet_name=None)

# Now, df_dict is a dictionary where each sheet name is a key, and each value is a DataFrame
# Example: Access each sheet by its name
df_sheet1 = df_dict['Iko']
df_sheet2 = df_dict['Ikhza']
df_sheet3 = df_dict['bagus']
df_sheet4 = df_dict['(JKT) Sep - Des 23']
df_sheet5 = df_dict['(SBY) Juli - Des 23']
```

```python
# Iterate over each sheet and display its columns
for sheet_name, df in df_dict.items():
    print(f"\nColumns in sheet: {sheet_name}")
    print(df.columns.tolist())
```

```python
# Iterate over each sheet and display column data types
for sheet_name, df in df_dict.items():
    print(f"\nData types in sheet: {sheet_name}")
    print(df.dtypes)  # Shows each column and its data type

# Reset the display options if needed
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')
```

```python
for sheet_name, df in df_dict.items():
    print(f"\nData types in sheet: {sheet_name}")
    print(df.dtypes)  # Shows each column and its data type
```

```python
for sheet_name, df in df_dict.items():
    print(f"\nStatistics for sheet: {sheet_name}")
    print(df.describe())  # Only numerical columns are described by default
```

```python
for sheet_name, df in df_dict.items():
    print(f"\nUnique values count in sheet: {sheet_name}")
    print(df.nunique())
```

```python
for sheet_name, df in df_dict.items():
    print(f"\nInconsistent data types in sheet: {sheet_name}")
    for column in df.columns:
        if df[column].dtype == 'object':
            # Check if any non-numeric values are in columns that should be numeric
            try:
                pd.to_numeric(df[column])
            except ValueError:
                print(f"Inconsistent values found in column: {column}")
```
